```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as ndi
from skimage import io, color
```

```python
img = io.imread('imagen.jpeg')
img_gray = color.rgb2gray(img)
```

```python
img_gray.shape, img.shape
```

```python
plt.imshow(img_gray)
plt.axis('off')
plt.show()
```

```python
kernel = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])
```

```python
img_edges = ndi.convolve(img_gray, kernel)

fig = plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(img_gray, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Convolución')
plt.imshow(img_edges, cmap='gray')
plt.axis('off')

plt.show()
```

```python
kernel2 = np.array([
    [-1, -1, -1],
    [0, 0, 0],
    [1, 1, 1]
])

img_edges2 = ndi.convolve(img_gray, kernel2)

fig = plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(img_gray, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Convolución 2')
plt.imshow(img_edges2, cmap='gray')
plt.axis('off')

plt.show()
```
