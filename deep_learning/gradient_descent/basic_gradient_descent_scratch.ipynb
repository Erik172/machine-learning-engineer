```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from plotly import graph_objects as go
```

```python
sns.set_style('darkgrid')
plt.rcParams['figure.dpi'] = 80
```

```python
function = lambda theha: (theha - 5) ** 2

_x = np.linspace(-10, 15, 100)

plt.figure(figsize=(10, 6))
sns.lineplot(x=_x, y=function(_x))
plt.title('Function')
plt.show()
```

## Stochastic Gradient Descent (SGD)

```python
gradient = lambda theha: 2 * (theha - 5)

theha = -10
learning_rate = 0.05
epochs = 100

history = np.array([theha])

for _ in range(epochs):
    theha = theha - learning_rate * gradient(theha)
    history = np.append(history, theha)
    
plt.figure(figsize=(10, 6))
sns.lineplot(x=_x, y=function(_x))
plt.scatter(history, function(history), color='red', edgecolor='black', linewidth=.7, alpha=.7, s=70, label='Gradient Descent')
plt.title('Gradient Descent')
plt.legend()
plt.show()
```

## Momentum

```python
theha = -10
history = np.array([theha])
learning_rate = 0.05
epochs = 100
mu = 0.8
v = np.zeros_like(theha)

for _ in range(epochs):
    v = mu * v + (1 + mu) * gradient(theha)
    theha = theha - learning_rate * v
    history = np.append(history, [theha], axis=0)
    
plt.figure(figsize=(10, 6))
sns.lineplot(x=_x, y=function(_x))
plt.scatter(history, function(history), color='red', edgecolor='black', linewidth=.7, alpha=.7, s=70, label='Momentum')
plt.title('Momentum')
plt.legend()
```
