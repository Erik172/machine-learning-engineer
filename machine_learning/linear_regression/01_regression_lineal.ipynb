# Gradient Descent for Intercept and Slope

Formule for the gradient descent intercept: $-\frac{2}{N}\sum_{i=1}^N(y_i - (mx_i + b))$ 

O 

$\frac{1}{N}\sum_{i=1}^N((mx_i + b) - y_i)$

````python
def get_gradient_at_b(x, y, m, b):
    diff = 0

    for i in range(len(x)):
        y_val = y[i]
        x_val = x[i]
        # diff += (((m * x_val) + b) - y_val)
        diff += (y_val - ((m * x_val) + b))

    b_gradient = -2/len(x) * diff
    # b_gradient = 1 / len(x) * diff

    return b_gradient
````

Formule for the gradient descent slope: $\frac{1}{N}\sum_{i=1}^Nx_i((mx_i + b) - y_i)$

OR 

$-\frac{2}{N}\sum_{i=1}^Nx_i(y_i - (mx_i + b))$

````python
def get_gradient_at_m(x, y, m, b):
    diff = 0

    for i in range(len(x)):
        y_val = y[i]
        x_val = x[i]
        # diff += x_val * (((m * x_val) + b) - y_val)
        diff += x_val * (y_val - ((m * x_val) + b))

    m_gradient = -2/len(x) * diff
    # m_gradient = 1 / len(x) * diff

    return m_gradient
````

Ahora que sabemos cmo calcular el gradiente, queremos dar un "step" en esa direccin. Sin embargo, es importante pensar si ese paso es demasiado grande o demasiado pequeo. No queremos sobrepasar el error mnimo!

Podemos escalar el tamao del paso multiplicando el gradiente por una tasa de aprendizaje .

Para encontrar un nuevo `bvalor`, diramos:


```
new_b = current_b - (learning_rate * b_gradient)
```

donde current_best nuestra suposicin de cul `b` es el valor, `b_gradientes` el gradiente de la curva de prdida en nuestra suposicin actual y `learning_ratees` proporcional al tamao del paso que queremos dar.

````python
def step_gradient(x, y, b_current, m_current, learning_rate):
    b_gradient = get_gradient_at_b(x, y, m_current, b_current)
    m_gradient = get_gradient_at_m(x, y, m_current, b_current)

    b = b_current - (learning_rate * b_gradient)
    m = m_current - (learning_rate * m_gradient)

    return [b, m]
````

````python
# current intercept guess:
b = 0
# current slope guess:
m = 0

# Call your function here to update b and m
b, m = step_gradient(months, revenue, b, m, 0.01)

print(b, m)
````

````python
plt.figure(figsize=(10, 5))

sns.scatterplot(
    x=months,
    y=revenue,
    s=100,
    edgecolor='black',
    linewidth=.7,
    alpha=.7
)

sns.lineplot(
    x=months,
    y=[m * x + b for x in months],
    color='red'
)

plt.title("Sandra's Lemonade Stand")
plt.xlabel('Months')
plt.ylabel('Revenue (in millions)')
plt.show()
````

````python
def gradient_descent(x, y, learning_rate, num_iterations):
    b = 0
    m = 0

    for i in range(num_iterations):
        b, m = step_gradient(x, y, b, m, learning_rate)

    return b, m
````

````python
b, m = gradient_descent(months, revenue, 0.01, 1000)
print(f'b: {b}, m: {m}')
````

````python
y = [m * x + b for x in months]
````

````python
plt.figure(figsize=(10, 5))

sns.scatterplot(
    x=months,
    y=revenue,
    s=100,
    edgecolor='black',
    linewidth=.7,
    alpha=.7
)

sns.lineplot(
    x=months,
    y=y,
    color='red'
)

plt.title("Sandra's Lemonade Stand")
plt.xlabel('Months')
plt.ylabel('Revenue (in millions)')
plt.show()
````
