```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
temperature = np.array(range(60, 100, 2))
temperature[:5]
```

```python
temperature = temperature.reshape(-1, 1)
temperature[:5]
```

```python
sales = [65, 58, 46, 45, 44, 42, 40, 40, 36, 38, 38, 28, 30, 22, 27, 25, 25, 20, 15, 5]
```

```python
line = LinearRegression()
line.fit(temperature, sales)
```

```python
sales_pred = line.predict(temperature)
```

```python
plt.figure(figsize=(8, 5))
plt.title('Sales vs Temperature')
plt.scatter(temperature, sales)
plt.plot(temperature, sales_pred, color='red')
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.show()
```

```python
print('Sales = {:.2f} * Temperature + {:.2f}'.format(line.coef_[0], line.intercept_))
```

```python
mean_squared_error(sales, sales_pred)
```

```python
r2_score(sales, sales_pred)
```
